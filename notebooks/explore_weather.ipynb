```python
import pandas as pd 
import numpy as np
from scipy.stats import pearsonr

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
weather1 = pd.read_csv("../data/weather/weather_data1.csv")
weather2 = pd.read_csv("../data/weather/weather_data2.csv")
```

```python
# TODO: concatenate your data files together and save to a new csv file ("../data/weather/weather_data.csv")
weather = pd.concat([weather1, weather2], ignore_index=True)

#Inspect the dataset to ensure it is consistent, correctly formatted, and contains only "Minas Gerais" subdivision
weather['year'].min(), "to", weather['year'].max()
print(weather['subdivision'].unique())

# Save the concatenated data to a new CSV file
weather.to_csv("../data/weather/weather_data.csv", index=False)

```

```python
# TODO: view first 5 rows of your data
weather.head()
```

```python
# TODO: describe your numeric columns
weather.describe().T
```

```python
# TODO: display the null values in your dataframe
weather.isnull().sum()
```

```python
# TODO: BEGIN YOUR EDA!

# Univariate Analysis #

# Average Temperature #

# Plot the distribution of the temperature across the years for Minas Gerais
plt.figure(figsize=(10, 6))
sns.lineplot(data=weather, x="year", y="temp_avg", hue="subdivision", marker="o")
plt.title("Average Temperature (C) in Minas Gerais Over the Years [2006 - 2023] from Jan to May")
plt.xlabel("Year")
plt.ylabel("Average Temperature (C)")
plt.xticks(rotation=45)
plt.grid()
plt.legend(title="Subdivision")
plt.tight_layout()
plt.show()

# Calculate the descriptive statistics for the average temperature
temp_avg_stats = weather['temp_avg'].describe()
print("Descriptive Statistics for Average Temperature (C):")
print(temp_avg_stats)

```

```python
# TODO: BEGIN YOUR EDA!

#Univariate Analysis

# Minimum Humidity #

# Plot the distribution of the minimum humidity across the years for Minas Gerais
plt.figure(figsize=(10, 6))
sns.lineplot(data=weather, x="year", y="hum_min", hue="subdivision", marker="o", palette="Set2")
plt.title("Minimum Humidity (%) in Minas Gerais Over the Years [2006 - 2023] from Jan to May")
plt.xlabel("Year")
plt.ylabel("Minimum Humidity (%)")
plt.xticks(rotation=45)
plt.grid()
plt.legend(title="Subdivision")
plt.tight_layout()
plt.show()

# Calculate the descriptive statistics for the minimum humidity
hum_min_stats = weather['hum_min'].describe()
print("Descriptive Statistics for Minimum Humidity (%):")
print(hum_min_stats)
```

```python
# TODO: BEGIN YOUR EDA!

#Univariate Analysis

# Average Wind Speed #

# Plot the distribution of the average wind speed across the years for Minas Gerais
plt.figure(figsize=(10, 6))
sns.lineplot(data=weather, x="year", y="wind_avg", hue="subdivision", marker="o", palette="Set1")
plt.title("Average Wind Speed (m/s) in Minas Gerais Over the Years [2006 - 2023] from Jan to May")
plt.xlabel("Year")
plt.ylabel("Average Wind Speed (m/s)")
plt.xticks(rotation=45)
plt.grid()
plt.legend(title="Subdivision")
plt.tight_layout()
plt.show()

# Calculate the descriptive statistics for the average wind speed
wind_avg_stats = weather['wind_avg'].describe()
print("Descriptive Statistics for Average Wind Speed (m/s):")
print(wind_avg_stats)
```

```python
# TODO: BEGIN YOUR EDA!

#Univariate Analysis

# Highest Maximum Record Rainfall #

#Plot a bar plot of the maximum rainfall across the years for Minas Gerais
plt.figure(figsize=(10, 6))
sns.barplot(data=weather, x="year", y="rain_max", hue="subdivision", palette="Set3")
plt.title("Maximum Rainfall (mm) in Minas Gerais Over the Years [2006 - 2023] from Jan to May")
plt.xlabel("Year")
plt.ylabel("Maximum Rainfall (mm)")
plt.xticks(rotation=45)
plt.grid()
plt.legend(title="Subdivision")
plt.tight_layout()
plt.show()

# Extract the year with the highest maximum rainfall
weather[weather["rain_max"] == weather["rain_max"].max()][["year", "rain_max"]]
```

```python
# TODO: BEGIN YOUR EDA!

#Bivariate Analysis

# Relationship between max rain and max humidity #

# Plot a scatter plot of the maximum rainfall against the maximum humidity for Minas Gerais
plt.figure(figsize=(10, 6)) 
sns.scatterplot(data=weather, x="rain_max", y="hum_max", hue="subdivision", palette="Set2")
plt.title("Maximum Rainfall (mm) vs Maximum Humidity (%) in Minas Gerais [2006 - 2023] from Jan to May")
plt.xlabel("Maximum Rainfall (mm)")
plt.ylabel("Maximum Humidity (%)")
plt.grid()
plt.legend(title="Subdivision")
plt.tight_layout()
plt.show()

# Calculate the pearson correlation and p-value between maximum rainfall and maximum humidity
r_value, p_value = pearsonr(weather["rain_max"], weather["hum_max"])
print(f"Pearson Correlation Coefficient between Maximum Rainfall and Maximum Humidity: {r_value:.3f}")
print(f"The p-value is: {p_value:.3f}")
```

```python
# TODO: BEGIN YOUR EDA!

#Bivariate Analysis

# Relationship between avg wind speed and avg temp #

# Plot a scatter plot of the maximum rainfall against the maximum humidity for Minas Gerais
plt.figure(figsize=(10, 6))
sns.scatterplot(data=weather, x="temp_avg", y="wind_avg", hue="subdivision", palette="Set1")
plt.title("Average Temperature (C) vs Average Wind Speed (m/s) in Minas Gerais [2006 - 2023] from Jan to May")
plt.xlabel("Average Temperature (C)")
plt.ylabel("Average Wind Speed (m/s)")
plt.grid()  
plt.legend(title="Subdivision")
plt.tight_layout()
plt.show()

# Calculate the pearson correlation and p-value between average temperature and average wind speed
r_value, p_value = pearsonr(weather["temp_avg"], weather["wind_avg"])
print(f"Pearson Correlation Coefficient between Average Temperature and Average Wind Speed: {r_value:.3f}")
print(f"The p-value is: {p_value:.3f}")

```
