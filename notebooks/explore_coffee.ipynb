```python
import pandas as pd 
import numpy as np
from scipy.stats import pearsonr


import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe().T
```

```python
# TODO: display the null values in your dataframe
coffee.isnull().sum()
```

```python
# TODO: drop your null values
coffee.dropna(inplace=True)

# Check to see if null values are dropped
coffee.isnull().sum()
```

```python
# TODO: Select only Minas Gerais

coffee = coffee[coffee["subdivision"] == "Minas Gerais"]

#Check the first 5 rows of the data
coffee.head()

# Reset the index
coffee.reset_index(drop=True, inplace=True)
coffee.head()
```

```python
# TODO: BEGIN YOUR EDA!

# Univariate Analysis # 

# Plot the bearing trees over the years 
plt.figure(figsize=(12, 6))
sns.lineplot(data=coffee, x="year", y="bearing_trees", hue="subdivision", palette="Set2", marker = "o")
plt.title("Bearing Trees Over the Years")
plt.xlabel("Year")
plt.ylabel("Bearing Trees")
plt.xticks(coffee["year"].unique())
plt.legend(title="Subdivision")
plt.show()

# Calculate the descriptive statistics for the bearing trees
bearing_trees_stats = coffee["bearing_trees"].describe()
print("Descriptive Statistics for bearing trees:")
print(bearing_trees_stats)

# Plot the bearing trees distribution
plt.figure(figsize=(12, 6))
sns.histplot(coffee["bearing_trees"], bins=20, kde=True)
plt.title("Distribution of Bearing Trees")
plt.xlabel("Bearing Trees")
plt.ylabel("Frequency")
plt.show()
```

```python
# TODO: BEGIN YOUR EDA!

# Univariate Analysis # 

# Plot the bearing hectares over the years for Minas Gerais
plt.figure(figsize=(12, 6))
sns.lineplot(data=coffee, x="year", y="bearing_hectares", hue="subdivision", palette="Set2", marker = "o")
plt.title("Bearing Hectares Over the Years")
plt.xlabel("Year")
plt.ylabel("Bearing Hectares")
plt.xticks(coffee["year"].unique())
plt.legend(title="Subdivision")
plt.show()

# Calculate the descriptive statistics for the bearing hectares
bearing_hectares_stats = coffee["bearing_hectares"].describe()
print("Descriptive Statistics for bearing hectares:")
print(bearing_hectares_stats)

# Plot the bearing hectares distribution
plt.figure(figsize=(12, 6))
sns.histplot(coffee["bearing_hectares"], bins=20, kde=True)
plt.title("Distribution of Bearing Hectares")
plt.xlabel("Bearing Hectares")
plt.ylabel("Frequency")
plt.show()
```

```python
# TODO: BEGIN YOUR EDA!

# Univariate Analysis # 

# Plot the output in 60kg bags over the years for Minas Gerais
plt.figure(figsize=(12, 6))
sns.lineplot(data=coffee, x="year", y="60kgs_bag", hue="subdivision", palette="Set2", marker = "o")
plt.title("Output in 60kg Bags Over the Years")
plt.xlabel("Year")  
plt.ylabel("Output in 60kg Bags")
plt.xticks(coffee["year"].unique())
plt.legend(title="Subdivision")
plt.show()

# Calculate the descriptive statistics for the output in 60kg bags
output_60kg_bags_stats = coffee["60kgs_bag"].describe()
print("Descriptive Statistics for output in 60kg bags:")
print(output_60kg_bags_stats)

# Plot the output in 60kg bags distribution
plt.figure(figsize=(12, 6))
sns.histplot(coffee["60kgs_bag"], bins=20, kde=True)     
plt.title("Distribution of Output in 60kg Bags")
plt.xlabel("Output in 60kg Bags")
plt.ylabel("Frequency")
plt.show()
```

```python
# TODO: BEGIN YOUR EDA!

#Bivariate Analysis

# Plot the relationship between bearing trees and output in 60kg bags
plt.figure(figsize=(12, 6))
sns.scatterplot(data=coffee, x="bearing_trees", y="60kgs_bag", hue="subdivision", palette="Set2")
plt.title("Bearing Trees vs Output in 60kg Bags")
plt.xlabel("Bearing Trees")
plt.ylabel("Output in 60kg Bags")
plt.legend(title="Subdivision")
plt.show()

# Calculate the correlation between bearing trees and output in 60kg bags
r_value, p_value = pearsonr(coffee["bearing_trees"], coffee["60kgs_bag"])
print(f"Pearson Correlation Coefficient between Bearing Trees and Output in 60kg Bags: {r_value:.3f}")
print(f"The p-value is: {p_value:.3f}")

```

```python
# TODO: BEGIN YOUR EDA!

#Bivariate Analysis

# Plot the relationship between bearing hectares and output in 60kg bags
plt.figure(figsize=(12, 6))
sns.scatterplot(data=coffee, x="bearing_hectares", y="60kgs_bag", hue="subdivision", palette="Set2")
plt.title("Bearing Hectares vs Output in 60kg Bags")
plt.xlabel("Bearing Hectares")
plt.ylabel("Output in 60kg Bags")
plt.legend(title="Subdivision")
plt.show()

# Calculate the correlation between bearing hectares and output in 60kg bags
r_value, p_value = pearsonr(coffee["bearing_hectares"], coffee["60kgs_bag"])
print(f"Pearson Correlation Coefficient between Bearing Hectares and Output in 60kg Bags: {r_value:.3f}")
print(f"The p-value is: {p_value:.3f}")  
```
